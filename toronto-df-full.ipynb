I'm importing csv file for Toronto locations and for Geocordinates to work them and combine together

```python
import pandas as pd
df = pd.read_csv('tor-loc.csv')
df.head()
```

I'm dropping not needed column

```python
df = df.drop(['Unnamed: 0'], axis=1)
df.head()
```

```python
import pandas as pd
df1 = pd.read_csv('Geospatial_Coordinates.csv')
df1.head()
```

I'm renaming 'Postal Code' column as 'Postcode' which later will be used to merge two dataframes together

```python
df1 = df1.rename(columns={'Postal Code': 'Postcode'})
df1.head()
```

I'm adding geolocations to the postcode and neighborhoods by merging two dataframes with the "Postcode" being the key

```python
dft = pd.merge(df, df1, on='Postcode')
dft.head(10)
```

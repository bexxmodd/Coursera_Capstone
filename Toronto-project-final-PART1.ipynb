I import csv file after scrapping it from the https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M

```python
# import csv file after scrapping it from the wikipage
df = pd.read_csv('toronto.csv')
df.head()
```

I'm dropping 'Not assigned' value rows from Borough column

```python
df = df[df.Borough != 'Not assigned']
df.head(10)
```

```python
df.shape
```

I'm filling Neighborhood 'Not assigned' values from linked Borought column

```python

df = df.mask(df == 'Not assigned').ffill(1)
df.head(10)
```

I'm combining Neighborhoods based on Postcode

```python
df = df.groupby(['Postcode','Borough'])['Neighborhood'].apply(','.join).reset_index()
df.head(20)
```
